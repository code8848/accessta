```python
# Function to calculate network density
import math
import itertools
df_stations=pd.read_excel("C:/Users/P-Koirala/OneDrive - Texas A&M Transportation Institute/Desktop/Data/SUP/#RAW/Data/FullSUPIdentifiedData.xlsx", sheet_name="SUPData", usecols=["stationid", "Latitude",'Longitude'])
gdf_stations=gpd.GeoDataFrame(df_stations, geometry=gpd.points_from_xy(df_stations.Longitude, df_stations.Latitude), crs='EPSG:4326')
gdf_stations=gdf_stations.to_crs("EPSG:2277")
def network_density(stations, roadway_network, year_suffex):
    roadway_types = ['UI', 'UA', 'UC', 'UL', 'RI', 'RA', 'RC', 'RL']
    buffer_radii = [1320, 2640, 5280] #in feet
    buffer_radii_name = ['025mile','05mile','1mile']
    columns = ['stationid'] + [f'{roadway_type}_{buffer_radius}_{year_suffex}' for roadway_type, buffer_radius in itertools.product(roadway_types, buffer_radii_name )]
    df = pd.DataFrame(columns=columns)
    #Joining station and roadway network
    for buffer_radius, buffer_radius_name in zip(buffer_radii, buffer_radii_name):
        buffered_points = stations.copy()
        buffered_points.geometry = stations.geometry.buffer(buffer_radius)
        
        roadway_points = gpd.sjoin(roadway_network, buffered_points, how='inner', op='within')
        #Calculating Density of roadway (wrt type) for each location wrt the current buffer_radius
        density_by_station_and_type = {}
        ADT_by_station_and_type = {}
        for stationid, group in roadway_points.groupby(['stationid']):
            for roadway_type, roadway_group in group.groupby(['RU_F_SYSTE']):
                total_length = roadway_group['geometry'].length.sum()
                point_area = math.pi*buffer_radius**2
                density = (total_length / point_area).astype(float)
                density_by_station_and_type[(stationid, roadway_type)] = density
              
        #Storing density columnwise
        for (stationid, roadway_type), density in density_by_station_and_type.items():
            if stationid not in df.index:
                df = df.append(pd.DataFrame({'stationid': stationid}, index=[stationid]))
                
            df.at[stationid, f'{roadway_type}_{buffer_radius_name}_{year_suffex}'] = density
    return df
```

```python
#Function to calculate ADT and max speed
import math
import itertools
df_stations=pd.read_excel("C:/Users/P-Koirala/OneDrive - Texas A&M Transportation Institute/Desktop/Data/SUP/#RAW/Data/FullSUPIdentifiedData.xlsx", sheet_name="SUPData", usecols=["stationid", "Latitude",'Longitude'])
gdf_stations=gpd.GeoDataFrame(df_stations, geometry=gpd.points_from_xy(df_stations.Longitude, df_stations.Latitude), crs='EPSG:4326')
gdf_stations=gdf_stations.to_crs("EPSG:2277")

def network_var_density(stations, roadway_network, year_suffex):
    roadway_types = ['UI', 'UA', 'UC', 'UL', 'RI', 'RA', 'RC', 'RL']
    buffer_radii = [1320, 2640, 5280]
    buffer_radii_name = ['025mile','05mile','1mile']
    var=['ADT_CUR', 'ADT_previous_year'] #maxspd is seperately done
    columns = ['stationid'] + [f'{vari}_{roadway_type}_{buffer_radius}_{year_suffex}' for vari, roadway_type, buffer_radius in itertools.product(var,roadway_types, buffer_radii_name )] + [f'MaxSpd_{buffer_radius}_{year_suffex}' for buffer_radius in  buffer_radii_name]
    #columns2 = ['stationid'] + [f'{var}_{roadway_type}_{buffer_radius}_{year_suffex}' for roadway_type, buffer_radius in itertools.product(var,roadway_types, buffer_radii_name )]
    temp_df = pd.DataFrame(columns=columns)
    
    #Joining station and roadway network
    for buffer_radius, buffer_radius_name in zip(buffer_radii, buffer_radii_name):
        buffered_points = stations.copy()
        buffered_points.geometry = stations.geometry.buffer(buffer_radius)
        
        roadway_points = gpd.sjoin(roadway_network, buffered_points, how='inner', op='within')
        #Calculating Density
       
        ADT_by_station_and_type = {}
        ADT_prev_year_by_station_and_type={}
        Numlanes_by_station_and_type={}
        MaxSpd_by_station={}
        intersection={}
        for stationid, group in roadway_points.groupby(['stationid']):
            MaxSpd_by_station[(stationid)] = group['SPD_MAX'].max()
            for roadway_type, roadway_group in group.groupby(['RU_F_SYSTE']):
                ADT_by_station_and_type[(stationid, roadway_type)] = roadway_group['ADT_CUR'].mean()
                ADT_prev_year_by_station_and_type[(stationid, roadway_type)] = roadway_group['HY_1'].mean()
                #Numlanes_by_station_and_type[(stationid, roadway_type)] = roadway_group['NUM_LANES'].max()
            #intersection[stationid] = group['NBR_SGNL'].sum() + group['NBR_STOP_S'].sum() + group['ATGRD_OTHR'].sum()
        #store the above values in the dataframe
        temp_df = pd.DataFrame(columns=columns)
        for (stationid, roadway_type), adt in ADT_by_station_and_type.items():
            if stationid not in temp_df.index:
                temp_df = temp_df.append(pd.DataFrame({'stationid': stationid}, index=[stationid]))
            temp_df.at[stationid, f'ADT_CUR_{roadway_type}_{buffer_radius_name}_{year_suffex}'] = adt
            temp_df.at[stationid, f'ADT_previous_year_{roadway_type}_{buffer_radius_name}_{year_suffex}'] = ADT_prev_year_by_station_and_type[(stationid, roadway_type)]
            temp_df.at[stationid, f'MaxSpd_{buffer_radius_name}_{year_suffex}'] = MaxSpd_by_station[(stationid)]


    return temp_df

      
```

```python
#Executing the above functions and joining the two dataframes for different years
shape_files=[shp_road16, shp_road17,shp_road18,shp_road19,shp_road20,shp_road21 ] #different years shp files are used from here
suffex=['16','17','18','19','20','21']
network_density_df=network_density(gdf_stations,shp_road15,year_suffex='15' )
network_var_df=network_var_density(gdf_stations,shp_road15,year_suffex='15')
network_var_df=network_var_df.loc[:, network_var_df.columns != 'stationid']
DF=pd.concat([network_density_df, network_var_df], axis=1)
for roadway_network,year_suffex in zip(shape_files, suffex):
     print(year_suffex)
     network_density_=network_density(gdf_stations,roadway_network,year_suffex )
     network_density_df=network_density_.loc[:, network_density_.columns != 'stationid']
     network_var_df=network_var_density(gdf_stations,roadway_network,year_suffex)
     network_var_df=network_var_df.loc[:, network_var_df.columns != 'stationid']
     DF=pd.concat([DF,network_density_df, network_var_df], axis=1)
```
